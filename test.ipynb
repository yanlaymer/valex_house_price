```python
# parse only digits from levels column
df.level.fillna('0', inplace=True)

def extract_digits(x):
    if isinstance(x, int):
        return int(x)
    else:
        # parse digit from string
        return int(''.join(filter(str.isdigit, x)))
    
df['level'] = df['level'].apply(extract_digits)
```

```python
def square_to_rooms(square):
    # create logic to convert square to rooms
    return square // 30
```

```python
df['rooms'] = df['square'].apply(square_to_rooms)
```

```python
# assign list value for household_buildings column
df['household_buildings'] = df['household_buildings'].apply(lambda x: [x])
```

```python
import numpy as np
df['internet'] = np.random.choice(['оптика', 'проводной'], size=len(df))
```

```python
df['local_phone'] = np.random.choice(['есть', 'есть возможность подключение'], size=len(df))
```

```python
df.material.value_counts(dropna=False)
```

```python
materials = ['деревянный', 'иное', 'каркасно-камышитовый', 'каркасно-щитовой', 'кирпичный', 'монолитный', 'панельный', 'пеноблочный', 'сэндвич-панели', 'шлакоблочный']
def check_materials(x):
    if x == 'кирпич':
        return 'кирпичный'
    elif x == 'саман':
        return 'каркасно-камышитовый'
    elif x == 'шлакоблок':
        return 'шлакоблочный'
    elif x == 'газоблок' or x == 'пеноблок' or x == 'полистирол-бетон':
        return 'пеноблочный'
    else:
        return np.random.choice(materials)
    
df['material'] = df['material'].apply(check_materials)

df.material.value_counts()
```

```python
df['security_system'] = np.random.choice(['да', 'нет'], size=len(df))
```

```python
df['household_number'] = np.random.choice([1, 2, 3, 4, 5], size=len(df))
```

```python
df.gas_system.value_counts()

df['gas_system'] = 'магистральный'
df['bathroom'] = np.random.choice(['раздельный', 'совмещенный', '2 с/у и более'], size=len(df))
df['water'] = 'центральная'
```

```python
df['house_number'] = df['house_number'].astype('str')
```

```python
from src.predictor import HouseModel

cols = list(HouseModel.__dict__['__annotations__'].keys())
```

```python
df.condition.value_counts()

def condition_check(condition):
    if condition == 'удовлетворительное':
        return 'среднее'
    elif condition.replace(' ', '') == 'хорошее':
        return 'хорошее'
    else:
        return condition
    
df.condition = df.condition.apply(condition_check)
```

```python
for row, frame in df.iterrows():
    house = HouseModel(**frame[cols].to_dict())
    predictor = HousePricePredictor()
    df.at[row, 'predicted_price'] = np.round(predictor.predict_price(house), -4)
    df.at[row, 'price_per_square_meter'] = np.round(predictor.price_per_square_meter, -2)
```

```python
df.to_excel('valex_house_predicted.xlsx', index=False)
```
